```python
import numpy as np 

A = np.array(((1,-1),(1,-1.00001)))
A2 = np.array(((1, -1),(1, -0.99999)))

b = np.array((1, 0))

print("Questão a)\n")

x = np.linalg.solve(A, b)
det = np.linalg.det(A)
cond = np.linalg.cond(A)

x2 = np.linalg.solve(A2, b)
det2 = np.linalg.det(A2)
cond2 = np.linalg.cond(A2)

print(f"x = {x}")
print(f"determinante de x = {det}")
print(f"condicionamento de x = {cond}")

print("___________\n")

print(f"x2 = {x2}")
print(f"determinante de x2 = {det2}")
print(f"condicionamento de x2 = {cond2}")

print("\n\n\nQuestão b)\n")

def calccond(n):
    matriz_b = np.zeros((n, n), dtype=float)
    b = np.zeros((n, 1), dtype=float)
    b.fill(1)

    for i in range(n):
        for j in range(n):
            matriz_b[i][j] = round(1 / (i + 1 + j + 1 - 1), 5)
        #print(f"matriz b: {matriz_b[i]}")
        #print("\n")

    print(f"matriz b | com n = {n}: \n{matriz_b}")
    condb = np.linalg.cond(matriz_b)
    print(f"\ndeterminate matriz_b: {np.linalg.det(matriz_b)}")
    print(f"\nsolução matriz_b: {np.linalg.solve(matriz_b, b)}")
    print(f"\ncondicionamento matriz_b: {condb}")

for n in range(3, 11):
    calccond(n)
    print("\n---\n")
```
